### **Programa 2: Maximo, Minimo, Rango y Dispersion de datos**

```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
Fab1=[]
Fab2=[]
for i in range(100):
  num_aleatorio= round(random.uniform(50,100),2)
  Fab1.append(num_aleatorio)

print(Fab1)

#

for i in range(100):
  num_aleatorio= round(random.uniform(50,100),2)
  Fab2.append(num_aleatorio)

print(Fab2)
```

```python
maximo1=max(Fab1)
minimo1=min(Fab1)
rango1=maximo1-minimo1
media1=mean(Fab1)
mediana1=median(Fab1)
moda1=mode(Fab1)
desviacion_estandarFab1=stdev(Fab1)

print("Estadistica desriptiva de la Fabrica 1\n", "Media=", media1, "\n Mediana= ", mediana1, "\n Moda= ", moda1, "\n Desviacion Estandar= ", desviacion_estandarFab1)
#

maximo2=max(Fab2)
minimo2=min(Fab2)
rango2=maximo1-minimo1
media2=mean(Fab2)
mediana2=median(Fab2)
moda2=mode(Fab2)
desviacion_estandarFab2=stdev(Fab2)

print("Estadistica desriptiva de la Fabrica 2\n", "Media=", media2, "\n Mediana= ", mediana2, "\n Moda= ", moda2, "\n Desviacion Estandar= ", desviacion_estandarFab2)
```

```python
plt.figure(figsize = (8,6))
plt.hist(Fab1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - desviacion_estandarFab1, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={desviacion_estandarFab1:.2f}')
plt.axvline(media1 + desviacion_estandarFab1, color='black', linestyle='dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()

plt.figure(figsize = (8,6))
plt.hist(Fab1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'Media={media2:.2f}')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana2:.2f}')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda2:.2f}')
plt.axvline(media2 - desviacion_estandarFab2, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={desviacion_estandarFab2:.2f}')
plt.axvline(media2 + desviacion_estandarFab2, color='black', linestyle='dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 2")
plt.legend()

plt.show()
```
